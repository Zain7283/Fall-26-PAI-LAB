# Lab 1 - Task 1
## Kaggle Competition: House Price Prediction

### Objective:
Build a supervised machine learning regression model to predict house prices.


## Step 1: Import Required Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
```

## Step 2: Load Dataset
Make sure train.csv is in the same folder as this notebook.

```python
data = pd.read_csv('train.csv')
data.head()
```

## Step 3: Select Target Variable (y) and Features (X)

```python
y = data['SalePrice']

features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr']
X = data[features]

X.head()
```

## Step 4: Split Dataset (70% Training, 30% Testing)

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=1
)

print('Training Data Size:', X_train.shape)
print('Testing Data Size:', X_test.shape)
```

## Step 5: Create and Train the Model
We use Decision Tree Regressor.

```python
model = DecisionTreeRegressor(random_state=1)
model.fit(X_train, y_train)
```

## Step 6: Make Predictions

```python
predictions = model.predict(X_test)
predictions[:5]
```

## Step 7: Evaluate Model Performance
Using Mean Absolute Error (MAE).

```python
mae = mean_absolute_error(y_test, predictions)
print('Mean Absolute Error:', mae)
```

## Conclusion

This task is a supervised regression problem because:
- We have input features (X)
- We have known output values (SalePrice)
- The model learns the relationship between inputs and output

The model performance is evaluated using Mean Absolute Error.
