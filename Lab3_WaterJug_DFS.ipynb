# Lab 3 - Water Jug Problem using DFS

This program solves the water jug problem using Depth First Search. The goal is to measure exactly 2 liters using a 4 liter jug and a 3 liter jug.

```python
CAPACITY_JUG1 = 4
CAPACITY_JUG2 = 3
GOAL = 2

visited = set()

def dfs(x, y, path):
    if (x, y) in visited:
        return False

    visited.add((x, y))
    path.append((x, y))

    if x == GOAL or y == GOAL:
        print('Goal Reached')
        print('Solution Path:')
        for step in path:
            print(step)
        return True

    moves = [
        (CAPACITY_JUG1, y),
        (x, CAPACITY_JUG2),
        (0, y),
        (x, 0),
        (x - min(x, CAPACITY_JUG2 - y), y + min(x, CAPACITY_JUG2 - y)),
        (x + min(y, CAPACITY_JUG1 - x), y - min(y, CAPACITY_JUG1 - x))
    ]

    for move in moves:
        print((x, y), '->', move)
        if dfs(move[0], move[1], path.copy()):
            return True

    return False

dfs(0, 0, [])
```
